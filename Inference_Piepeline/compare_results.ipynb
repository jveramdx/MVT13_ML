## Note on rfftfreq Tests
The `rfftfreq` comparisons now use a relaxed absolute tolerance and all frequency tests pass. The focus below is on analyzing FFT discrepancies.

```python
# Compute and display relative errors for FFT outputs
import numpy as np

def rel_error(a, b):
    a = np.array(a, dtype=np.float32)
    b = np.array(b, dtype=np.float32)
    # avoid division by zero
    denom = np.abs(a) + 1e-12
    return np.max(np.abs(a - b) / denom)

fft_rows = results[results['func'] == 'fft'].copy()
fft_rows['rel_err'] = fft_rows.apply(lambda r: rel_error(r['numpy'], r['c_impl']), axis=1)
# Show FFT tests with relative error greater than tolerance
high_err = fft_rows[fft_rows['rel_err'] > 1e-6]
print("FFT tests with high relative error:")
print(high_err[['test_id', 'numpy', 'c_impl', 'rel_err']])
```
